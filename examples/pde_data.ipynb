## Ginzburg-Landau Reaction Diffusion PDE


$$ \partial_t u = \partial_{xx} u + \partial_{xx} v + u (1- (u^2+v^2)) + 1.5 (u^2+v^2) v $$
$$ \partial_t v = \partial_{xx} u + \partial_{xx} v + v (1- (u^2+v^2)) - 1.5 (u^2+v^2) u $$

See http://basilisk.fr/src/examples/ginzburg-landau.c

```python
u_data=np.load(os.path.join(PDEConfig.ginzburg_dir, '256', 'Ar_256_0_05.npy'))
v_data=np.load(os.path.join(PDEConfig.ginzburg_dir, '256', 'Ai_256_0_05.npy'))

#norm u^2 + v^2
uv_data=np.load(os.path.join(PDEConfig.ginzburg_dir, '256', 'A2_256_0_05.npy'))
```

```python
print(u_data.shape, v_data.shape, uv_data.shape)
```

```python
fig, ax = plt.subplots(8,8, figsize=(10,10))

for i,z in enumerate(ax.ravel()):
    #z.pcolormesh(u_data[i*10].reshape(*coord_dims), cmap='RdBu', shading='gouraud')
    z.pcolormesh(u_data[i*10].reshape(u_data.shape[1:]), cmap='RdYlBu')#, shading='gouraud')
    z.axis('off')
```

```python
fig, ax = plt.subplots(8,8, figsize=(10,10))

for i,z in enumerate(ax.ravel()):
    #z.pcolormesh(u_data[i*10].reshape(*coord_dims), cmap='RdBu', shading='gouraud')
    z.pcolormesh(uv_data[i*10].reshape(uv_data.shape[1:]), cmap='RdYlBu')#, shading='gouraud')
    z.axis('off')
```

## Simple Spiral Reaction Diffusion PDE


$$ \partial_t u = 0.1(\partial_{xx} u + \partial_{xx} v) + u (1- (u^2+v^2)) +  (u^2+v^2) v $$
$$ \partial_t v = 0.1(\partial_{xx} u + \partial_{xx} v) + v (1- (u^2+v^2)) -  (u^2+v^2) u $$

See: https://pysindy.readthedocs.io/en/latest/examples/10_PDEFIND_examples/example.html

```python

u_data=np.load(os.path.join(PDEConfig.sindpy_data,'gen', 'rdiff2d_u.npy'))
v_data=np.load(os.path.join(PDEConfig.sindpy_data,'gen', 'rdiff2d_v.npy'))
```

```python
fig, ax = plt.subplots(3,3, figsize=(10,10))

for i,z in enumerate(ax.ravel()):
    #z.pcolormesh(u_data[i*10].reshape(*coord_dims), cmap='RdBu', shading='gouraud')
    z.pcolormesh(u_data[i*4].reshape(u_data.shape[1:]), cmap='RdYlBu')#, shading='gouraud')
    z.axis('off')
```
